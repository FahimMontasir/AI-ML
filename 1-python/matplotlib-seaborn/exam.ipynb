### Question : 1

```python
import numpy as np
import pandas as pd

df = pd.read_csv('final-employee-ds.csv')
training_hours = df['TrainingHours'].values

training_hours = training_hours.astype(float)

mean_value = np.mean(training_hours)
std_value = np.std(training_hours)

print(mean_value)
print(std_value)
```

#### Question : 2

```python

df['Salary'] = df.groupby('Department')['Salary'].transform(lambda x: x.fillna(x.median()))

filtered_df = df[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5)]

total_salary = filtered_df['Salary'].sum()

print(total_salary)
```

#### Question : 3

```python


filtered_df = df[
    (df['Department'].isin(['IT', 'Finance'])) &
    (df['Age'] >= 30) & (df['Age'] <= 45) &
    (df['PerformanceScore'] > 88)
]

filtered_df = filtered_df.sort_values(by='Salary', ascending=False)

result = filtered_df[['Name', 'Department', 'Salary', 'PerformanceScore']].reset_index(drop=True)

print(result)
```

#### Question : 4

```python

df['SalaryPerHour'] = df['Salary'] / (df['WorkHoursPerWeek'] * 4)

top_5 = df.nlargest(5, 'SalaryPerHour')

result = top_5[['Name', 'SalaryPerHour']]

print(result)
```

#### Question : 5

```python
import plotly.express as px

gender_dept_counts = df.groupby(['Department', 'Gender']).size().reset_index(name='Count')


fig = px.bar(gender_dept_counts, 
             x='Department', 
             y='Count', 
             color='Gender',
             barmode='group',
             color_discrete_map={'Male': '#3498db', 'Female': '#e74c3c'},
             title='Employee Count by Department and Gender',
             labels={'Count': 'Number of Employees'})


fig.update_layout(
    xaxis_title='Department',
    yaxis_title='Number of Employees',
    font=dict(size=12),
    title_font=dict(size=14, family='Arial Black'),
    legend_title_text='Gender'
)


fig.show()
```
